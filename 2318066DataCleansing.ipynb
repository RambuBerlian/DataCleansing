# **4. Data Cleansing-Standarisasi Kolom ID Formasi**

Algoritma yang akan digunakan:

1. Filter data hanya untuk formasi_id yang panjangnya 36 karakter.

2. Buat salinan data hasil filter agar indeks bisa di-reset (reset_index).

3. Hapus duplikasi berdasarkan formasi_id agar setiap ID unik.

4. Tampilkan jumlah data awal, jumlah data setelah dibersihkan, dan contoh beberapa baris.

```python
# Filter hanya data dengan formasi_id valid (36 karakter)
data_clean = data[data['formasi_id'].str.len() == 36].copy()

# Reset index agar rapi
data_clean.reset_index(drop=True, inplace=True)
```

```python
# Tampilkan ringkasan jumlah data
print("Jumlah data awal  :", len(data))
print("Jumlah data bersih:", len(data_clean))

# Tampilkan data kolom formasi_id
from IPython.display import display
display(data_clean[['formasi_id']].head(15))
```
